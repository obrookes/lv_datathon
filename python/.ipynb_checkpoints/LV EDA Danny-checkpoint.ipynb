## Data
The data are already pre-processed, and have been split into `train` and `val`, for training and validation respectively, in a separate file. We load them in with `pandas` here.

```python
train = pd.read_csv("data/train.csv")
val   = pd.read_csv("data/val.csv")
```

We can look at what features there are.

```python
print(train.columns)
```

Some are irrelevant, such as `State` and `State Code` being the same. `Customer` is an ID variable which is not necessary, and all values in `Country` are USA. We should also drop `Gender` to remove bias based on sex.

```python
train = train.drop(columns = ["Customer", "Country", "State Code", "Gender"])
```

From inspection we can see that we have the following continuous features

```python
cont = ["Income", "Monthly Premium Auto", "Months Since Last Claim", 
        "Months Since Policy Inception", "Number of Open Complaints",
        "Number of Policies", "Effective To Date"]
print(cont)
```

And the following categorical

```python
cat = list(train.columns.drop(cont).drop("Total Claim Amount"))
print(cat)
```

Also the `Effective to Date` column is in varied date formats, so we can convert them as differences (in days) from January 1st 2011.

```python
def convert_dates(x):
    return (parser.parse(x) - parser.parse("01/01/2011")).days
train["Effective To Date"] = train["Effective To Date"].apply(convert_dates)
```

```python
train[cat].describe(include = "all").T
```

```python
train[cont].describe(include = "all").T
```

## Plot continuous variables

Plotting the continuous variables in `cont` against each other should show some correlations.

```python
train_melt = pd.melt(train[np.hstack(("Total Claim Amount", cont))], id_vars = "Total Claim Amount")

g = sns.FacetGrid(train_melt, col="variable", col_wrap = 3, sharey="col", sharex = None);
g.map(sns.scatterplot, "value", "Total Claim Amount");
```

- Linear effect between `Total Claim Amount` (response) and `Monthly Premium Auto`, with heteroscedasticity, so that the variance increases as does the Monthly Premium. 
- Negative linear trend on `Income`, although it is not as obvious, and only seems to affect the 'max' values.
- Possibly a quadratic trend with `Number of Open Complaints`
- Others don't seem to be interesting.

## Some basic models

### Setup
First, define some metrics

```python
def MAE(train_preds, val_preds, y, y_val):
    error_train = np.mean(abs(train_preds - y))
    error_val   = np.mean(abs(val_preds - y_val))

    # Print MAE
    print("Training MAE: {}".format(np.round(error_train, 3)))
    print("Validation MAE: {}".format(np.round(error_val, 3)))    
    
def RMSE(train_preds, val_preds, y, y_val):
    error_train = np.sqrt(np.mean(np.power(train_preds - y, 2)))
    error_val   = np.sqrt(np.mean(np.power(val_preds - y_val, 2)))

    # Print MAE
    print("Training RMSE: {}".format(np.round(error_train, 3)))
    print("Validation RMSE: {}".format(np.round(error_val, 3)))    
```

Format the data (using a different file).


```python
import python.data_funs as df

train = pd.read_csv("data/train.csv")
val   = pd.read_csv("data/val.csv")

X, y = df.set_up(train)
X_val, y_val = df.set_up(val)

X = df.onehot(X)
X_val = df.onehot(X_val)
```

### Random forest

```python
import sklearn.ensemble as ske
from time import time

start = time()
rf = ske.RandomForestRegressor(n_estimators = 100,
                               verbose = False, 
                               n_jobs = 4,
                               random_state = 1)

rf.fit(X, y);
elapsed = time() - start
print("Elapsed run time for RF: {} seconds".format(np.round(elapsed, 5)))

MAE(rf.predict(X), rf.predict(X_val), y, y_val)
RMSE(rf.predict(X), rf.predict(X_val), y, y_val)
```

Very low training MAE but validation MAE high, perhaps an overfit. 

### Elastic net regularisation

```python
import sklearn.linear_model as lm
en = lm.ElasticNet(random_state=1)
en.fit(X, y)
MAE(en.predict(X), en.predict(X_val), y, y_val)
RMSE(en.predict(X), en.predict(X_val), y, y_val)
```

Also not great validation MAE, although somehow better than random forest.
